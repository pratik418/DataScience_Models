```python
print("""Explanation:
->Identifies missing values in several physiological variables and the target (NSP)
->Missing values account for a small percentage of the dataset""")
df.isnull().sum()
```

```python
print("""Explanation:
->The target variable (NSP) represents fetal health status
->Imputing the target could introduce incorrect labels
->Rows with missing NSP are removed to preserve label integrity""")
df = df.dropna(subset=['NSP'])
df
```

```python
print("""Explanation:
->Median imputation is used because the data contains outliers
->The median is more robust than the mean for skewed medical data
->Ensures no remaining missing values""")
num_cols = df.columns.drop('NSP')

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
print("""Explanation:
->NSP should only contain three valid classes:
1 = Normal
2 = Suspect
3 = Pathological
->Floating-point noise and invalid labels were corrected by rounding
->Invalid classes were removed
->Converted to integer for categorical interpretation""")
df['NSP'] = df['NSP'].round()
```

```python
df = df[df['NSP'].isin([1, 2, 3])]
```

```python
df['NSP'] = df['NSP'].astype(int)
```

```python
print("""Explanation:
->The IQR method identifies extreme values
->Many outliers are present
->Outliers are retained because they represent clinically significant fetal conditions
1. Baseline Fetal Heart Rate (LB) – 10 outliers
Represents unusually low or high fetal heart rates
Extreme heart rates may indicate fetal distress or abnormal physiological conditions
Low outlier count suggests LB is generally stable
2. Accelerations (AC) – 41 outliers
Accelerations reflect fetal responsiveness
Very high acceleration counts are uncommon but can occur during increased fetal activity
These outliers likely represent bursts of movement rather than errors
3. Fetal Movements (FM) – 340 outliers
FM contains many zero values with occasional high spikes
This zero-inflated nature leads to a large number of IQR-based outliers
High FM values indicate periods of intense fetal activity and are clinically meaningful
4. Uterine Contractions (UC) – 13 outliers
UC generally falls within a narrow range
Extreme contraction frequencies are rare and may indicate labor or stress
Small number of outliers reflects expected physiological behavior
5. Late Decelerations (DL) – 122 outliers
Late decelerations are abnormal and associated with placental insufficiency
Their rarity causes extreme values to be flagged as outliers
These values are clinically important indicators of fetal compromise
6. Short Decelerations (DS) – 120 outliers
Short decelerations are uncommon but relevant
High counts indicate frequent transient heart rate drops
Outliers represent abnormal fetal response patterns
7. Prolonged Decelerations (DP) – 283 outliers
Prolonged decelerations are rare but serious events
Their low baseline frequency leads to many IQR-identified outliers
These observations are critical for identifying pathological cases
8. Abnormal Short-Term Variability (ASTV) – 10 outliers
ASTV measures the percentage of abnormal variability
Extreme values strongly correlate with fetal distress
Small outlier count indicates stability with clear pathological extremes
9. Mean Short-Term Variability (MSTV) – 77 outliers
Low or unusually high MSTV values indicate abnormal autonomic regulation
Outliers highlight compromised fetal heart rate control
10. Abnormal Long-Term Variability (ALTV) – 317 outliers
ALTV is a strong indicator of pathological fetal conditions
High outlier count reflects the presence of distressed fetuses
These values are essential for distinguishing fetal states
11. Mean Long-Term Variability (MLTV) – 80 outliers
MLTV extremes indicate abnormal heart rate oscillations
Outliers represent physiologically significant deviations
12. Width of FHR Histogram (Width) – 10 outliers
Width reflects overall heart rate variability
Extreme widths are associated with abnormal patterns
Limited outliers suggest stable behavior for most observations
13. Tendency – 10 outliers
Tendency captures overall FHR trends
Extreme upward or downward trends may indicate fetal stress
Low outlier count suggests consistency in measurement""")
features = df.drop(columns=['NSP'])

Q1 = features.quantile(0.25)
Q3 = features.quantile(0.75)
IQR = Q3 - Q1

outliers = ((features < (Q1 - 1.5 * IQR)) |
            (features > (Q3 + 1.5 * IQR))).sum()

outliers
```

```python
columns = df.drop(columns='NSP').columns
def skew_type(skew):
    if skew > 0.5:
        return "Right-skewed"
    elif skew < -0.5:
        return "Left-skewed"
    else:
        return "Approximately symmetric"


fig, axes = plt.subplots(
    nrows=int(np.ceil(len(columns) / 4)),
    ncols=4,
    figsize=(18, 14)
)

axes = axes.flatten()


for ax, col in zip(axes, columns):
    sns.histplot(
        df[col],
        bins=30,
        kde=True,          # ✅ KDE added
        edgecolor='black',
        ax=ax
    )
    
    skew_val = df[col].skew()
    skew_label = skew_type(skew_val)
    
    ax.text(
        0.95, 0.95,
        f"Skewness = {skew_val:.2f}\n{skew_label}",
        transform=ax.transAxes,
        ha='right',
        va='top',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black")
    )
    
    ax.set_title(col)
    ax.set_xlabel("")
    ax.set_ylabel("")


for ax in axes[len(columns):]:
    ax.remove()

plt.suptitle(
    "Distribution of Cardiotocographic Features\n(Histogram with KDE and Skewness)",
    fontsize=18
)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
print("""Explanation:
Variables: LB, AC, FM, UC
Target: NSP
This pair plot explores relationships between:
Baseline fetal heart rate (LB)
Accelerations (AC)
Fetal movements (FM)
Uterine contractions (UC)
The diagonal KDE plots show the distribution of each variable, while the off-diagonal scatter plots show pairwise relationships.
Observations
1)Baseline Fetal Heart Rate (LB)
LB is tightly clustered within a normal physiological range
Minimal separation between fetal states
Indicates LB alone is not a strong discriminator of fetal condition
2)Accelerations (AC)
Most values are close to zero
Occasional higher values represent active fetal responses
Slight clustering of pathological cases at lower AC values
3) Fetal Movements (FM)
Highly right-skewed with many near-zero values
Pathological cases show fewer and weaker movements
Confirms FM as a sensitive but sparse indicator
4)Uterine Contractions (UC)
Narrow distribution across all classes
Weak visual correlation with fetal state
Suggests UC affects fetal health indirectly""")
sample_df = df.sample(300, random_state=42)
group1 = ['LB', 'AC', 'FM', 'UC', 'NSP']
group2 = ['ASTV', 'MSTV', 'ALTV', 'MLTV', 'NSP']
group3 = ['DL', 'DS', 'DP', 'Width', 'Tendency', 'NSP']

groups = {
    "group1_baseline_activity": group1,
    "group2_variability": group2,
    "group3_decelerations": group3
}

for name, group in groups.items():
    pair = sns.pairplot(
        sample_df[group],
        hue='NSP',
        diag_kind='kde',
        height=3.5,
        corner=True
    )
    
    pair.fig.suptitle(
        f"Pair Plot: {name.replace('_', ' ').title()}",
        y=1.02,
        fontsize=16
    )

    pair.savefig(
        f"{name}_pairplot.png",
        dpi=300,
        bbox_inches='tight'
    )
    
    plt.show()
```
