```python
import numpy as np
from scipy import stats


data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])


n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  

confidence_level = 0.99
alpha = 1 - confidence_level

print("Sample size:", n)
print("Sample mean:", round(sample_mean, 4))
print("Sample standard deviation:", round(sample_std, 4))


# Part (a): 99% Confidence Interval using t-distribution

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)


standard_error_t = sample_std / np.sqrt(n)


margin_error_t = t_critical * standard_error_t

# confidence interval
ci_t = (
    sample_mean - margin_error_t,
    sample_mean + margin_error_t
)

print("\n99% Confidence Interval using t-distribution:")
print("Lower bound:", round(ci_t[0], 4))
print("Upper bound:", round(ci_t[1], 4))


# Part (b): 99% Confidence Interval using known sigma


sigma = 0.2  


z_critical = stats.norm.ppf(1 - alpha/2)


standard_error_z = sigma / np.sqrt(n)


margin_error_z = z_critical * standard_error_z

# confidence interval
ci_z = (
    sample_mean - margin_error_z,
    sample_mean + margin_error_z
)

print("\n99% Confidence Interval using z-distribution:")
print("Lower bound:", round(ci_z[0], 4))
print("Upper bound:", round(ci_z[1], 4))
```
