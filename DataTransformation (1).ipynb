```python
df.replace("?", np.nan, inplace=True)

print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("\nNumerical Columns:", list(num_cols))
print("\nCategorical Columns:", list(cat_cols))
```

```python
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)


for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df.isnull().sum())
```

```python
standard_scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[num_cols] = standard_scaler.fit_transform(df[num_cols])
```

```python
minmax_scaler = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[num_cols] = minmax_scaler.fit_transform(df[num_cols])
```

```python
print("""Several numerical features in the Adult dataset, such as age, education-num, and hours-per-week, follow approximately symmetric or moderately skewed distributions. These features are not bounded to a fixed range and represent continuous measurements.
When using linear models (Logistic Regression, Linear SVM) to predict income.
When applying distance-based algorithms where differences from the mean are important.
When performing dimensionality reduction techniques such as PCA.
In the Adult dataset:
Age values range roughly between 17 and 90.
Hours-per-week typically cluster around 40.
Standard scaling centers these features around zero, allowing the model to learn relationships without one feature dominating due to scale.
Standard scaling preserves the underlying distribution of these features while ensuring that no single numerical attribute disproportionately influences the model.

2. Min-Max Scaling 
Some numerical features in the Adult dataset, such as capital-gain and capital-loss, have extremely large ranges and are highly skewed. These features can negatively affect models that expect inputs within a limited range.
When training neural networks to predict income.
When using K-Nearest Neighbors or K-Means clustering.
When a bounded input range is required for faster and more stable training.
In the Adult dataset:
Capital-gain values range from 0 to very large numbers.
Min-Max scaling compresses these values into the [0,1] range, preventing extreme values from dominating distance calculations.
Min-Max scaling maintains relative differences while ensuring numerical stability for models sensitive to feature magnitude.

In the Adult Census Income dataset, Standard Scaling is preferred for features such as age, education-num, and hours-per-week because these variables are continuous and moderately distributed, making them suitable for models that rely on variance and distance from the mean. Min-Max Scaling is preferred for highly skewed features such as capital-gain and capital-loss, especially when training neural networks or distance-based models, as it constrains values within a fixed range and improves numerical stability.""")
```

```python
low_cardinality = [col for col in cat_cols if df[col].nunique() < 5]
high_cardinality = [col for col in cat_cols if df[col].nunique() >= 5]

print("\nLow Cardinality Columns (<5 categories):", low_cardinality)
print("\nHigh Cardinality Columns (>=5 categories):", high_cardinality)
print("""We divided categorical features into:
Low-cardinality (< 5 categories)
High-cardinality (≥ 5 categories)""")
```

```python
df_ohe = pd.get_dummies(
    df,
    columns=low_cardinality,
    drop_first=True
)
print("""Applied One-Hot Encoding to categorical variables with fewer than 5 categories.
No ordinal relationship between categories
Prevents false ranking
Ideal for small category sets
Converts each category into a binary feature (0 or 1).
Pros
No assumption of order between categories
Safe for linear and distance-based models
Prevents ordinal bias
Cons
Increases number of features
Not suitable for high-cardinality variables
Can increase memory usage""")
```

```python
df_encoded = df_ohe.copy()
label_encoder = LabelEncoder()

for col in high_cardinality:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
print("""Applied Label Encoding to variables with more than 5 categories.
One-Hot Encoding would create too many columns
Label Encoding is memory-efficient
Suitable for tree-based models
Maps each category to a unique numerical value.
Pros
Memory efficient
Works well for tree-based models
Simple and fast
Cons
Introduces false ordering
Not suitable for linear models unless handled carefully""")
```

```python
#Feature 1: Work Intensity
df_encoded['work_intensity'] = (
    df_encoded['hours_per_week'] / (df_encoded['age'] + 1)
)

# Feature 2: Education per Age
df_encoded['education_per_age'] = (
    df_encoded['education_num'] / (df_encoded['age'] + 1)
)
print(""""Created a feature combining age and working hours.
Raw features may not fully capture behavior.
This feature represents effort relative to age.
Provides deeper insight into work patterns linked to income.
Feature 2: Education per Age
Created a feature combining education level and age.
Helps identify individuals who achieved higher education earlier.
Improves the model’s ability to differentiate income levels.
""")
```

```python
df_encoded['capital_gain_log'] = np.log1p(df_encoded['capital_gain'])
print("""Transformation of Skewed Numerical Feature
Applied a log transformation to the capital-gain feature.
Capital gain is highly skewed
Extreme values distort model learning
Log transformation stabilizes variance
Reduces skewness
Makes distribution more normal
Improves model performance""")
```

```python
X = df_encoded.drop('income_>50K', axis=1, errors='ignore')

# Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)

iso_forest.fit(X)

# Get anomaly scores
anomaly_scores = iso_forest.decision_function(X)

# Add anomaly score to dataset
df_encoded['anomaly_score'] = anomaly_scores
print("""Detects outliers and rare patterns
Helps identify features contributing to anomalies
Improves robustness of feature selection""")
```

```python
import ppscore as pps
pps_scores = []

for col in df_encoded.columns:
    if col != 'income_>50K':
        score = pps.score(df_encoded, col, 'income_>50K')
        pps_scores.append(score)

pps_df = pd.DataFrame(pps_scores)
pps_df = pps_df.sort_values(by='ppscore', ascending=False)
```
