```python
import math
from scipy.stats import norm
```

```python
sample_mean = 3050
mu_0 = 1000 + 5 * 600   
sigma = 5 * 25         
n = 25
alpha = 0.05


z = (sample_mean - mu_0) / (sigma / math.sqrt(n))


z_critical = norm.ppf(1 - alpha)


p_value = 1 - norm.cdf(z)


print("Hypothesis Test for Weekly Operating Costs")
print("------------------------------------------")
print(f"Theoretical Mean (μ₀): {mu_0}")
print(f"Sample Mean (x̄): {sample_mean}")
print(f"Standard Deviation (σ): {sigma}")
print(f"Sample Size (n): {n}")
print(f"Z-statistic: {z:.2f}")
print(f"Critical Z-value (α = 0.05): {z_critical:.3f}")
print(f"P-value: {p_value:.5f}")

if z > z_critical:
    print("Decision: Reject the null hypothesis.")
    print("Conclusion: There is evidence that weekly costs are higher than the model suggests.")
else:
    print("Decision: Fail to reject the null hypothesis.")
    print("Conclusion: There is no evidence that weekly costs are higher than the model suggests.")
```
